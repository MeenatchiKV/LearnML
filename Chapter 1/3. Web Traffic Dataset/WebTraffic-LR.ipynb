# LINEAR REGRESSION ON WEB TRAFFIC DATA SET

Given a web traffic dataset with two colums x and y.
where x represents the months in numbers and y represents the traffic experienced by a particular website on that particular month.

Importing necessary python packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

from sklearn import preprocessing, model_selection
from sklearn.linear_model import LinearRegression

style.use('ggplot')
```

Reading two datasets one for training and the other for testing.

dataframe.dropna(inplace=True) drops all the rows that does not contain any value.
inplace=True parameter makes the changes to the variable, otherwise changes will not be committed.

```python
train = pd.read_csv('dataset/web_traffic.tsv',sep='\t')
test = pd.read_csv('dataset/web_test.tsv', sep='\t')
train.dropna(inplace=True)
test.dropna(inplace=True)

```

seperating the features from the labels and reshaping the numpy arrays.

```python
x_train = np.array(train['x']).reshape(-1,1)
y_train = np.array(train['y']).reshape(-1,1)
x_test = np.array(test['x']).reshape(-1,1)
y_test = np.array(test['y']).reshape(-1,1)

```

Visualization of the  training and testing dataset.

```python
plt.scatter(x_train,y_train)
plt.show()
plt.scatter(x_test,y_test)
plt.show()
```

Creating a Linear Regression model to fin in the training data and printing the accuracy

```python
clf = LinearRegression()
clf.fit(x_train,y_train)

acc = clf.score(x_test,y_test)

print(acc)
```
